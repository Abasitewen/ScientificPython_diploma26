### **Exercise 1:**
Write a function to solve a quadratic equation. Use the formula with the discriminant, only find the real solutions, if they don't exist print some message.

Hint: for the square root just use the power 0.5 to avoid having to call a special function which we haven't seen yet.

```python
def quad_eqtn(a,b,c):
    discriminant = b**2 - 4*a*c

    if discriminant >0:
        x_sol1 = (-b+ discriminant**0.5)/ (2*a)
        x_sol2 = (-b-discriminant**0.5)/(2*a)
        print(' solutions to the quadratic equation are:')
        return x_sol1,x_sol2
    
    elif discriminant == 0:
        print('The solution to the quadratic equation is:')
        x_sol = -b/(2*a)
        return x_sol
    else:
        print('No real solutions exist')
        return none

quad_eqtn(2,-10,12)
    
    
```

### **Exercise 2:**
The Recamán sequence is a famous sequence invented by the Colombian mathematician, Bernardo Recamán Santos. It is defined by the following algorithm, starting at $a_0=0$:

$ a_n=a_{n-1}−n$ if $a_{n-1}-n > 0  $ and has not already appeared in the sequence; 

$ a_n=a_{n−1}+n$ otherwise

Write a function that returns the first N elements of this sequence.

You can use lists or sets to store the numbers.

(to check that it works - you should get 0, 1, 3, 6, 2, 7, 13, 20, 12, 21, 11, 22, 10, 23, 9, 24, 8, 25, 43, 62, 42, 63, 41, 18, 42, 17, 43, 16, 44, 15, 45, 14, 46, 79, 113, 78, 114, 77, 39, 78, 38, 79, 37, 80, 36, 81, 35, 82)

```python
def recaman_sequence(N):
    sequence = []
    seen = set()
    current = 0
    sequence.append(current)
    seen.add(current)

    for n in range(1, N):
        prev = sequence[-1]
        candidate = prev - n
        if candidate > 0 and candidate not in seen:
            current = candidate
        else:
            current = prev + n
        sequence.append(current)
        seen.add(current)
    
    return sequence
```

```python
print(recaman_sequence(10))
```

### **Exercise 5:**
Create a function that returns a list of all pairs of factors (as tuples) of a given number using list comprehension.

```python
def pair_factors(n):
    list1 = [(i,n//i) for i in range(1,n) if n % i==0]
    return list1
```

### **Exercise 6:**
Create a list of all numbers divisible by 3 and smaller than 1000 in 4 different ways:
 
 - naive way where you just create a list and append data in a loop
 - using list comprehension
 - slicing an existing list
 - directly with `range()`

Time all of them.

```python
import time
print("Finding all numbers divisible by 3 and less than 1000")
print()

print("METHOD 1: Simple loop with if statement")
start_time = time.time()

numbers1 = []
for i in range(1000):
    if i % 3 == 0:  # Check if divisible by 3
        numbers1.append(i)

end_time = time.time()
time1 = end_time - start_time
print(f"Time taken: {time1:.6f} seconds")
print(f"Found {len(numbers1)} numbers")
print()


print("METHOD 2: List comprehension")
start_time = time.time()

numbers2 = [i for i in range(1000) if i % 3 == 0]

end_time = time.time()
time2 = end_time - start_time
print(f"Time taken: {time2:.6f} seconds")
print(f"Found {len(numbers2)} numbers")
print()


print("METHOD 3: Slicing (take every 3rd number)")
start_time = time.time()

all_numbers = list(range(1000))
numbers3 = all_numbers[::3]  # Start at 0, take every 3rd number

end_time = time.time()
time3 = end_time - start_time
print(f"Time taken: {time3:.6f} seconds")
print(f"Found {len(numbers3)} numbers")
print()


print("METHOD 4: Direct range (start=0, stop=1000, step=3)")
start_time = time.time()

numbers4 = list(range(0, 1000, 3))

end_time = time.time()
time4 = end_time - start_time
print(f"Time taken: {time4:.6f} seconds")
print(f"Found {len(numbers4)} numbers")
print()


print("RESULTS:")
print(f"First 10 numbers: {numbers4[:10]}")
print(f"Last 10 numbers: {numbers4[-10:]}")
print()

print("TIMING COMPARISON:")
print(f"Method 1 (Simple loop):      {time1:.6f} seconds")
print(f"Method 2 (List comprehension): {time2:.6f} seconds")
print(f"Method 3 (Slicing):          {time3:.6f} seconds")
print(f"Method 4 (Direct range):     {time4:.6f} seconds")
print()


if time4 < time1 and time4 < time2 and time4 < time3:
    print("Winner: Method 4 (Direct range) is the FASTEST!")
```

Exercise 7
Write a function to check if a string (containing spaces and capital letters, but no punctuation marks) is a palindrome (use help to see which functions are available for a string). 


```python
def is_palindrome(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove spaces
    text = text.replace(" ", "")
    
    # Reverse the string
    reversed_text = text[::-1]
    
    # Check if both are the same
    if text == reversed_text:
        return True
    else:
        return False
```

```python
# Example tests
print(is_palindrome("Madam")) 
print(is_palindrome("Never Odd Or Even"))
print(is_palindrome("Hello World"))
```

### **Exercise 8:**
Write a function to find a character with a maximum number of occurences in a string. Return a tuple with that character and the number. You can ignore ties. Only use string's inbuilt functions and the function `max` (use help to see how to use it).

```python
def max_occurrence(text):
    # Find the character with the maximum count
    max_char = max(text, key=text.count)
    
    # Count how many times it appears
    max_count = text.count(max_char)
    
    return (max_char, max_count)
```

```python
# Example tests
print(max_occurrence("banana"))     
print(max_occurrence("mississippi"))  
print(max_occurrence("hello"))
```

### **Exercise 9:**
Write a function that return the first N prime numbers using the Sieve of Eratosphenes algorithm

https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes

```python
def first_n_primes(n):
    if n <= 0:
        return []

    # Step 1: Estimate an upper limit for n primes
    limit = n * 15  

    # Step 2: Create a sieve (True means "prime")
    sieve = [True] * (limit + 1)
    sieve[0] = False
    sieve[1] = False

    # Step 3: Apply the Sieve of Eratosthenes
    for i in range(2, int(limit ** 0.5) + 1):
        if sieve[i]:
            for multiple in range(i * i, limit + 1, i):
                sieve[multiple] = False

    # Step 4: Collect primes
    primes = []
    for i in range(2, limit + 1):
        if sieve[i]:
            primes.append(i)
        if len(primes) == n:
            break

    return primes
```

```python
# Example tests
print(first_n_primes(5))    
print(first_n_primes(10))   
```
